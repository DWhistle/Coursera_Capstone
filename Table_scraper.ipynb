```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

```python
html = urllib.request.urlopen("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
bs = html.read()

html_code = bs.decode("utf8")
html.close()
```

```python
soup = BeautifulSoup(html_code)
```

```python
table = soup.find('table',attrs = {'class':'wikitable sortable'})
tab = table.find_all('a')
td = table.find_all('td')
col1 = []
col2 = []
col3 = []
i = 0
for elem in td:
    if i == 0:
        col1.append(elem.renderContents().replace('&b', ''))
    elif i == 1:
        col1.append(elem.get('td'))
    elif i == 2:
        col1.append(elem.get('td'))
        i = 0
print(col1)
final = []
for link in tab:
    final.append(link.get('title'))
```

```python
df = pd.DataFrame(final)
```
